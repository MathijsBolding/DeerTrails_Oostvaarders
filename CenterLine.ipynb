```python
import pygeoops 
import geopandas
import matplotlib as plt
```

```python
#load in the vector 
polygonTrails = geopandas.read_file("data/polygon_sample.gpkg")
```

```python
polygons = list(polygonTrails.geometry)
```

```python
centerlinesTrails = pygeoops.centerline(polygons)
centerline_gs = geopandas.GeoSeries(centerlinesTrails)
```

```python
centerline_gs.to_file("test_centerline.gpkg")
```

```python
# Plot the original polygon
polygonTrails.plot(color='lightblue', edgecolor='black', alpha=0.5)  # Plot polygon
centerline_gs.plot(color='red', linewidth=5)  # Plot centerline

# Add title and labels
plt.title("Polygon and Its Centerline")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
